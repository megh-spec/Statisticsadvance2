1.  What is hypothesis testing in statistics?
    - Hypothesis testing in statistics is a formal procedure using sample data to evaluate a claim (hypothesis) about a population, deciding if there's enough evidence to reject a starting assumption (null hypothesis) in favor of an alternative hypothesis. It helps determine if observed effects in a sample are real or just due to random chance, using concepts like p-values and significance levels (α) to make data-driven decisions, a core part of statistical inference.

2. What is the null hypothesis, and how does it differ from the alternative
hypothesis?
   - The null hypothesis (H₀) is a default statement of no effect or no difference between variables, assuming results are due to chance, while the alternative hypothesis (Hₐ or H₁) proposes there is a significant effect or relationship, which researchers aim to find evidence for by rejecting the null. The key difference is that H₀ suggests equality or no change (e.g., =), while H₁ suggests inequality or change (e.g., ≠, <, >), with statistical tests designed to provide evidence to either reject H₀ (supporting H₁) or fail to reject H₀.

3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
   - The significance level, denoted by α (alpha), is the maximum probability of rejecting the null hypothesis (H₀) when it is actually true.

    In simple words, it is the risk we are willing to take of making a wrong decision.

Meaning of Significance Level -

It sets a cut-off point for decision-making.

It controls the chance of a Type I error (false rejection of H₀).

Commonly used values are:

α = 0.05 (5% risk)

α = 0.01 (1% risk)


Role in Deciding the Outcome of a Test -

Comparison with p-value :

After performing a test, we calculate the p-value.

Decision rule:

If p-value ≤ α → Reject H₀

If p-value > α → Fail to reject H₀

Determines Critical Region:

The significance level defines the critical region (rejection region).

If the test statistic falls in this region, H₀ is rejected.

Controls Errors:

Lower α (e.g., 0.01) → Stricter test, less chance of false rejection.

Higher α (e.g., 0.10) → Less strict, more chance of false rejection.


4. What are Type I and Type II errors? Give examples of each.
   - In hypothesis testing, errors can occur while making decisions about the null hypothesis. These errors are of two types: Type I and Type II.

  1.Type I Error (α Error)

Definition:
A Type I error occurs when the null hypothesis (H₀) is rejected even though it is actually true.

It is also called a false positive.

Probability of Type I error = α (significance level).

Example:
H₀: A medicine has no effect.
Test result: We conclude the medicine works.
But in reality, it does not work.
→ This is a Type I error.

2.Type II Error (β Error)

Definition:
A Type II error occurs when the null hypothesis (H₀) is not rejected even though it is false.

It is also called a false negative.

Probability of Type II error = β.

Example:
H₀: A machine is working properly.
Test result: We conclude the machine is fine.
But in reality, it is faulty.
→ This is a Type II error.

5. What is the difference between a Z-test and a T-test? Explain when to use
each.
   - In hypothesis testing, the Z-test and T-test are both used to test claims about a population mean, but they are applied in different situations. A Z-test is used when the sample size is large (generally 30 or more) and the population standard deviation is known. In this case, the sampling distribution of the mean follows the normal distribution, and the Z statistic is used to make decisions. On the other hand, a T-test is used when the sample size is small (less than 30) and the population standard deviation is unknown. Here, the sample standard deviation is used as an estimate, and the test statistic follows the Student’s t-distribution, which accounts for extra variability due to small samples. In short, use a Z-test for large samples with known population variance, and use a T-test for small samples when the population variance is unknown.



```python
#6. Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.
# Python program to generate a binomial distribution and plot histogram

import random
import matplotlib.pyplot as plt

n = 10
p = 0.5
samples = 1000

binomial_data = []
for _ in range(samples):
    successes = 0
    for _ in range(n):
        if random.random() < p:
            successes += 1
    binomial_data.append(successes)

print("First 10 values of generated binomial data:")
print(binomial_data[:10])

plt.hist(binomial_data, bins=range(0, n+2), edgecolor='black')
plt.xlabel("Number of successes")
plt.ylabel("Frequency")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.show()

#7.  Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
# sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
# Implementing Z-test for hypothesis testing using the given sample data

import math
from statistics import mean
from scipy.stats import norm

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

mu0 = 50

sigma = 1

n = len(sample_data)

x_bar = mean(sample_data)

z = (x_bar - mu0) / (sigma / math.sqrt(n))

p_value = 2 * (1 - norm.cdf(abs(z)))

print("Sample Mean =", x_bar)
print("Z-statistic =", z)
print("P-value =", p_value)


#8. Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
# Simulate data from a normal distribution, calculate 95% confidence interval,
# and plot the data using Matplotlib

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(42)

mu = 50
sigma = 5
n = 100

data = np.random.normal(mu, sigma, n)

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

alpha = 0.05
z = norm.ppf(1 - alpha/2)
margin_error = z * (sample_std / np.sqrt(n))

ci_lower = sample_mean - margin_error
ci_upper = sample_mean + margin_error

print("Sample Mean =", round(sample_mean, 3))
print("95% Confidence Interval = (", round(ci_lower, 3), ",", round(ci_upper, 3), ")")

plt.hist(data, bins=15, edgecolor='black')
plt.axvline(sample_mean)
plt.axvline(ci_lower)
plt.axvline(ci_upper)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Normal Distribution Sample with 95% Confidence Interval")
plt.show()

#9.  Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.
# Python function to calculate Z-scores and visualize standardized data

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    z_scores = [(x - mean) / std for x in data]
    return z_scores, mean, std

data = [45, 50, 55, 60, 48, 52, 58, 62, 47, 53]

z_scores, mean, std = calculate_z_scores(data)

print("Mean =", round(mean, 2))
print("Standard Deviation =", round(std, 2))
print("Z-scores:")
for z in z_scores:
    print(round(z, 2))

plt.hist(z_scores, bins=8, edgecolor='black')
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.title("Histogram of Z-scores (Standardized Data)")
plt.show()
```
